```python
import pandas as pd
import matplotlib
import sklearn
```

```python
mvp = pd.read_csv("mvp.csv")

mvp.columns = mvp.iloc[0]

mvp = mvp[1:].reset_index(drop=True)
mvp
```

```python
mvp["Age"] = pd.to_numeric(mvp["Age"], errors="coerce")
mvp["G"] = pd.to_numeric(mvp["G"], errors="coerce")
mvp["MP"] = pd.to_numeric(mvp["MP"], errors="coerce")
mvp["PTS"] = pd.to_numeric(mvp["PTS"], errors="coerce")
mvp["TRB"] = pd.to_numeric(mvp["TRB"], errors="coerce")
mvp["AST"] = pd.to_numeric(mvp["AST"], errors="coerce")
mvp["STL"] = pd.to_numeric(mvp["STL"], errors="coerce")
mvp["BLK"] = pd.to_numeric(mvp["BLK"], errors="coerce")
mvp["FG%"] = pd.to_numeric(mvp["FG%"], errors="coerce")
mvp["3P%"] = pd.to_numeric(mvp["3P%"], errors="coerce")
mvp["FT%"] = pd.to_numeric(mvp["FT%"], errors="coerce")
mvp["WS"] = pd.to_numeric(mvp["WS"], errors="coerce")
mvp["WS/48"] = pd.to_numeric(mvp["WS/48"], errors="coerce")
mvp.dtypes
```
