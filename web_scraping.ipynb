```python
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

dfs = []

for year in years:
    with open(f"mvp/{year}.html") as f:
        page = f.read()

    soup = BeautifulSoup(page, "html.parser")

    # Ensure the header row exists before trying to remove it
    header_row = soup.find('tr', class_="over_header")
    if header_row:
        header_row.decompose()

    # Ensure table with id="mvp" exists
    mvp_table = soup.find(id="mvp")
    if mvp_table:
        mvp = pd.read_html(StringIO(str(mvp_table)))[0]
        dfs.append(mvp)
        mvp["Year"] = year
    else:
        print(f"Warning: No table found for year {year}")
```

```python
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

dfs = [] # Replace with your actual years list

for year in years:
    print(f"\rProcessing year: {year}", end="", flush=True)    
    with open(f"team/{year}.html") as f:
        page = f.read()

    soup = BeautifulSoup(page, "html.parser")

    # Remove all <th> elements inside <tr class="thead">, if present
    thead = soup.find('tr', class_='thead')
    if thead:
        for th in thead.find_all("th"):
            th.decompose()
        thead.decompose()

    # Process Eastern Conference
    team_table = soup.find(id="divs_standings_E")
    if team_table:
        team = pd.read_html(StringIO(str(team_table)))[0]
        team["Year"] = year
        if "Eastern Conference" in team.columns:
            team["Team"] = team["Eastern Conference"]
            del team["Eastern Conference"]
        dfs.append(team)

    # Process Western Conference
    team_table = soup.find(id="divs_standings_W")
    if team_table:
        team = pd.read_html(StringIO(str(team_table)))[0]
        team["Year"] = year
        if "Western Conference" in team.columns:
            team["Team"] = team["Western Conference"]
            del team["Western Conference"]
        dfs.append(team)

# Final DataFrame with all processed years
final_df = pd.concat(dfs, ignore_index=True)

# Display the DataFrame
import ace_tools as tools
tools.display_dataframe_to_user(name="Extracted Team Data", dataframe=final_df)
```
